```python
class Phone: # Base/parent class
  def __init__(self, brand, model_name, price):
    self.brand = brand
    self.model_name = model_name
    self.price = max(price, 0)
  @staticmethod
  def make_a_call(phone_num):
    print(f"Calling {phone_num}...")

  def full_name(self):
    return f"{self.brand} {self.model_name}"
```

```python
class SmartPhone(Phone): # Derived/Child Class
  def __init__(self, brand, model_name, price, ram, internal_memory, rare_camera):
    super().__init__(brand, model_name, price) # common way of doing inheritance
    # Phone.__init__(self, brand, model_name, price) # inherit (# Uncommon way of inheritance)
    self.ram = ram
    self.internal_memory = internal_memory
    self.rare_camera = rare_camera
```

```python
phone1 = Phone('Nokia', 1100, 4000)
```

```python
phone1.price
```

```python
phone1.full_name()
```
