```python
class Laptop: # class
  discount = 10 # class variable
  def __init__(self, name, price):
    self.name = name # class attributes
    self.price = price

  @classmethod
  def from_string(cls, string): # class method
      # Expected string: "my laptop is m6600 and price is 5000"
      import re
      name = re.search(r"is (\w+)", string).group(1)
      price = int(re.search(r"price is (\d+)", string).group(1))
      return cls(name, int(price))

  def applydiscount(self): # instance method
    discountAmount = (self.price/100)*self.discount
    final_price = self.price - discountAmount
    return final_price
```

```python
m6600 = Laptop('m6600', 5000)
```

```python
m6600.__dict__
```

```python
m6600.applydiscount()
```

```python
m4400 = Laptop.from_string("my laptop is m4400 and price is 3000")
```
