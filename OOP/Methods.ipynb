```python
class Student:
  def __init__(self, name,age,marks):
    self.name = name
    self.age = age
    self.marks = marks

  def result(self):
    if self.marks >= 500:
      return "Pass"
    else:
      return "Fail"
```

```python
s1 = Student('Kushal', 31, 500)
```

```python
s1.name
```

```python
s1.result()
```

Methods Exercise 1

```python
s1 = Student('Kushal', 31, 900)
s2 = Student('Pierre', 34, 200)
s3 = Student('Kevin', 30, 300)
s4 = Student('Blake', 29, 700)
s5 = Student('Sophia', 27, 600)
s6 = Student('Linda', 23, 400)
```

```python
student_objects = [s1, s2, s3, s4, s5, s6]
```

```python
pass_students = []
fail_students = []
students_result = {}
for student in student_objects:
  if student.result() == "Pass":
    pass_students.append(student.name)
    students_result['Pass Student'] = pass_students
  else:
    fail_students.append(student.name)
    students_result['Fail Student'] = fail_students
```

```python
fail_students
```

```python
students_result
```

```python
class Product:
  def __init__(self, product_name, price):
    self.product_name = product_name
    self.price = price

  def discount(self, discount_value):
    discount_amount = (self.price/100)*discount_value
    final_price = self.price - discount_amount
    return int(final_price)
```

```python
laptop = Product('Lenevo', 5500)
```

```python
laptop.discount(5)
```
